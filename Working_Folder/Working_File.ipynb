```python
# Filter the data 
drug_related = df6[df6['drugs_related_stop'] == True]

# Group by hour and count 
drug_activity_by_hour = drug_related.groupby('stop_time').size()

# Plot 
plt.figure(figsize=(24, 6))
drug_activity_by_hour.plot(kind='bar', color='red')
plt.title('Drug Activity by Time of Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Drug-Related Stops')
plt.locator_params(axis='x', nbins=24 )
# plt.xticks([x for x in range(0,24)])
plt.grid(axis='x')
plt.show()

#kinda hard to figure out the proper spacing for the x axis. Making do with locator_params
```
